```python
# Extract date_time_captured values from query results
# @params query_results a list of tuples that represent the rows returned from cursor.fetch*()
# @returns a list of datetime objects
def extract_date_time_captured(query_results: list) -> list:
    datetime_list = []
    for row in query_results:
        datetime_list.append(row[1])
    return datetime_list

# Connect to the PostgreSQL database
try:
    connection = psycopg2.connect(
        host=yaml_data['host'],
        port=yaml_data['port'],
        database=yaml_data['database'],
        user=yaml_data['user'],
        password=yaml_data['password']
    )
    
    # Create a cursor object
    cursor = connection.cursor()

    # Loop until reach end of photos.image table
    cursor.execute("SELECT COUNT(*) FROM photos.image")
    num_records_in_photos_table = cursor.fetchone()[0]

    # Extract filename and datetimes of photos.image
    query_get_data = f"SELECT nasa_filename, date_time_captured FROM photos.image\
                            ORDER BY nasa_filename, date_time_captured"
    cursor.execute(query_get_data)
    results = cursor.fetchall()
    list_date_time_captured = extract_date_time_captured(results)

    # Set variables
    offset = 0  # Tracks where we are in the original table
    movie_list = []  # Final list of time-lapse movies that will be found
    marginOfError = datetime.timedelta(seconds=1)  # Used to compare time deltas
    MAX_INT = float('inf')
    start = MAX_INT  # Will compare anytime we start to find a new cluster
    end = 0
    finished_cluster = False  # Indicator for determining last photo in cluster

    # Begin iterating thorugh list of datetimes
    while (offset < num_records_in_photos_table - 2):
        # Must check at least 3 entries to determine if there is a pattern in increments; use window method
        window = [list_date_time_captured[offset], list_date_time_captured[offset+1], list_date_time_captured[offset+2]]
        interval1 = window[1] - window[0]
        interval2 = window[2] - window[1]
        
        # If the difference in intervals is beyond our margin of error, then this is not a possible timelapse
        if (interval1 > interval2+marginOfError or interval1 < interval2-marginOfError):
            # End the current cluster if it exists
            if start < MAX_INT:
                end = offset+2  # Exclusive
                finished_cluster = True
        else:
            # We want small time deltas
            if (interval1 > marginOfError) or (interval2 > marginOfError):
                offset = offset + 1
                continue
            # Possible timelapse, so take note of beginning index and end index
            if offset < start: start = offset

        # Found a possible cluster
        if finished_cluster:
            # Not enough to be qualified as a time lapse movie
            if end - start > 100:
                cluster = [record[0] for record in results[start:end]]
                movie_list.append(cluster)
            # Reset relevant variables
            start = MAX_INT
            end = offset
            finished_cluster = False
        offset = offset + 1
       
    print(f'Number of movies found: {len(movie_list)}')
    for movie in movie_list:
        print(f'Frames in movie: {len(movie)}')
        print(f'First frame: {movie[0]}, Last Frame: {movie[len(movie)-1]}')

except Exception as e:
    print(f"An error occurred: {type(e)}: {e}")

finally:
    # Close the cursor and connection
    if cursor:
        cursor.close()
    if connection:
        connection.close()
    
```

```python
import requests
import os
mission = "ISS028"
filename = ""
for movie in movie_list:
    mission = movie[0].split("-")[0]
    first_frame = movie[0]
    last_frame = movie[-1]
    directory = f"{first_frame}_to_{last_frame}"
    for frame in movie:
        filename = f"{directory}/{frame}"
        url = f'https://windows-on-earth.sdsc.osn.xsede.org/{mission}/{frame}'
        response = requests.get(url)
        if response.status_code == 200:
            try:
                original_umask = os.umask(0)
                os.makedirs(os.path.dirname(filename), mode=0o744, exist_ok=True)
            finally:
                os.umask(original_umask)
            with open(f'{filename}', "wb") as file:
                file.write(response.content)
                print(f"File {filename} downloaded successfully!")
       else:
            print("Failed to download the file.")
```
